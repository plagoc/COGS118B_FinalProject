```python
# Import necessary packages that would allow us to convert them to text
import numpy as np
import pdfminer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
!nltk.download('stopwords')
!nltk.download('punkt')
```

```python
# For Windows
from os import listdir
from os.path import isfile, join

screenplayNameTxt = [f for f in listdir('MovieScreenplays') if isfile(join('MovieScreenplays', f))]
```

```python
screenplayNameTxt = screenplayNameTxt[1:]
```

```python
screenplayTxt = []

for name in screenplayNameTxt:
    test = open('MovieScreenplays/' + name, 'r', encoding="utf8")
    c = test.read()
    screenplayTxt.append(c)
```

```python
# Most work
print(screenplayTxt[0])
```

```python
# Remove stopwords
stop_words = set(stopwords.words('english'))
```

```python
word_tokens = word_tokenize(screenplayTxt[0])
```

```python
filtered_script = [w for w in word_tokens if not w in stop_words]
```

```python
print(filtered_script)
```

```python
filtered_scripts = []
for i in np.arange(len(screenplayTxt)):
    word_tokens = word_tokenize(screenplayTxt[i])
    filtered_scripts.append([w for w in word_tokens if not w in stop_words])
```

```python
len(filtered_scripts)
```

```python
filtered_scripts[0]
```

```python
outfile = open('sample1.txt' ,'w+', encoding="utf8")
for i in range(len(filtered_scripts[0])):
     outfile.write(filtered_scripts[0][i] + '\n')
outfile.close()
```
